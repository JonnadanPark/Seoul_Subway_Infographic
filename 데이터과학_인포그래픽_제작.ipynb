```python
import plotly.express as px
fig = px.histogram(x=df['승차총승객수'], # x축
                   y=df['노선명'], # y축
                   color=df['노선명'] , # dataframe에서 어떤 기준으로 색을 다르게 할 것이냐
                   histfunc="avg", # 히스토그램을 어떤 자료로 나타낼 것이냐? 이 자료에서는 평균(avg)로 나타낼 것이다
                   color_discrete_sequence=["#FF0000", # 1호선 = red
                                            "#1DDB16", # 2호선 = greend
                                            "#FFBB00", # 3호선 = yellow
                                            "#0100FF", # 4호선 = blue
                                            "#8324FF", # 5호선 = purple
                                            "#2ECCFA"], # 공황철도 = skyblue 
                   template="plotly_dark", # 배경 색
                   text_auto=True, # 텍스트를 넣는가?
                   title= "호선별 승차총승객수 평균" # 시각화 제목
                   )

fig.update_yaxes(categoryorder="total ascending") # 내림차순
fig.update_layout(height = 700, width = 800, hovermode = 'closest') # 시각화 화면 크기 설정
fig.update_traces(textfont_size=20, textangle=0, textposition="inside") # 텍스트의 크기, 각도, 위치
```
